## Overview of the DataSet

SpaceX has gained worldwide attention for a series of historic milestones. 

It is the only private company ever to return a spacecraft from low-earth orbit, which it first accomplished in December 2010.
SpaceX advertises Falcon 9 rocket launches on its website with a cost of 62 million dollars wheras other providers cost upward of 165 million dollars each, much of the savings is because Space X can reuse the first stage. 


Therefore if we can determine if the first stage will land, we can determine the cost of a launch. 

This information can be used if an alternate company wants to bid against SpaceX for a rocket launch.

This dataset includes a record for each payload carried during a SpaceX mission into outer space.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
%%sql
select * from spacex
createQuery = "create table INSTRUCTOR(ID INTEGER PRIMARY KEY NOT NULL, FNAME VARCHAR(20), LNAME VARCHAR(20), CITY VARCHAR(20), CCODE CHAR(2))"

createStmt = ibm_db.exec_immediate(conn,createQuery)
```

```python
#Construct the query that retrieves all rows from the INSTRUCTOR table
selectQuery = "select * from SPACEX"

#Execute the statement
selectStmt = ibm_db.exec_immediate(conn, selectQuery)
ibm_db.fetch_both(selectStmt)
```

```python
#Fetch the rest of the rows and print the ID and FNAME for those rows
while ibm_db.fetch_row(selectStmt) != False:
    print (" ID:",  ibm_db.result(selectStmt, 0), " FNAME:",  ibm_db.result(selectStmt, "FNAME"))
```

```python
import pandas
import ibm_db_dbi
```

```python
#connection for pandas
pconn = ibm_db_dbi.Connection(conn)
```

```python
#query statement to retrieve all rows in INSTRUCTOR table
selectQuery = "select distinct LAUNCH_SITE from SPACEX"

#retrieve the query results into a pandas dataframe
pdf = pandas.read_sql(selectQuery, pconn)

#print just the LNAME for first row in the pandas data frame
#pdf.DATE[0]
pdf
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
selectQuery = "select * from SPACEX WHERE LAUNCH_SITE LIKE 'CCA%' LIMIT 5"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
selectQuery = "select SUM(PAYLOAD_MASS__KG_) from SPACEX"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
selectQuery = "select AVG(PAYLOAD_MASS__KG_) from SPACEX WHERE BOOSTER_VERSION LIKE '%F9 v1.1%'"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
selectQuery = "select min(date) from SPACEX WHERE LANDING__OUTCOME LIKE '%Success%'"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
selectQuery = "select BOOSTER_VERSION from SPACEX WHERE LANDING__OUTCOME LIKE '%Success (drone%' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
selectQuery = "select COUNT(MISSION_OUTCOME) from SPACEX GROUP BY MISSION_OUTCOME"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
selectQuery = "SELECT BOOSTER_VERSION FROM SPACEX WHERE PAYLOAD_MASS__KG_ IN (select MAX(PAYLOAD_MASS__KG_) from SPACEX)"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```

### Task 9


##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
selectQuery = "SELECT BOOSTER_VERSION, LAUNCH_SITE FROM SPACEX WHERE LANDING__OUTCOME LIKE 'Failure (drone%'"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
selectQuery = "SELECT COUNT(DATE),LANDING__OUTCOME FROM SPACEX WHERE DATE >= '2010-06-04' AND DATE <= '2017-03-20' GROUP BY LANDING__OUTCOME ORDER BY LANDING__OUTCOME DESC"
pdf = pandas.read_sql(selectQuery, pconn)
#pdf.DATE[0]
pdf
```
